```python
# Renaming the SITE_CODE on Cell_geo to SITE_ID

cells_geo.rename(columns={'SITE_CODE': 'SITE_ID'}, inplace=True)
cells_geo.head()
```

```python
# Removing NULL from Cells_geo

cells_geo.dropna(inplace=True)
cells_geo.head()
```

```python
# Rename the Days dataset columns to be common
new_columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_NUMBER_INT', 
           'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']

first_day.columns = new_columns
second_day.columns = new_columns
third_day.columns = new_columns
```

```python
first_day.head(1)
```

```python
second_day.head(1)
```

```python
third_day.head(1)
```

```python
# Combining the Three CDR datasets

df_CDR = pd.concat([first_day, second_day, third_day])
df_CDR.head()
```

### Which ones were the most used city for the three days?

```python
# Checking the Villes whose status is in service
Villes = cells_geo[cells_geo['STATUS'] == 'In Service']
Villes
Villes['VILLES'].value_counts().head()
```

```python
# Dropping Unnecessary columns

df_CDR.drop(columns= ['CELL_ON_SITE','COUNTRY_A', 'COUNTRY_B'], inplace=True)
df_CDR.head()
```

```python
#Splitting the date_time column to create two columns 
df_CDR[['DATE', 'TIME']] = df_CDR.DATE_TIME.str.split(expand=True)
df_CDR
```

```python
# Dropping the Unnecessary DATE_TIME column

df_CDR.drop(columns=['DATE_TIME'], inplace=True)
df_CDR.head()
```

### Merging the Cells_geo dataset with the combined CDR dataset

```python
df = cells_geo.merge(df_CDR, on='SITE_ID')
df.head()
```

```python
# Grouping by Product and Time

df.groupby(['PRODUCT','TIME'])[['VALUE']].sum()
```
